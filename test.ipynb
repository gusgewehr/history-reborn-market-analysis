```python
import pandas as pd
from bs4 import BeautifulSoup
import requests

r = requests.get('https://www.historyreborn.net/?module=item&action=view&id=480149')

html_table = pd.read_html(r.text)
```

```python
html_table[1].set_index(0, inplace=True)
```

```python
soup = BeautifulSoup(r.text, 'html.parser')
```

```python
table = soup.find("table", {"id": "nova-sale-table"})
rows = []
ths = []

for tr in table.find_all("tr"):
    #ths = tr.find_all("th")
    #if ths != []:
    #    rows.append(ths)
    tds = tr.find_all("td")
    row = []
    for data in tds:
        try:
            row.append(data.text.rstrip().lstrip())
            row.append(data.find('a')['href'])
        except:
            row.append("")
    rows.append(row)

df = pd.DataFrame(rows)

df.columns = df.iloc[0]

df = df[1:]





    
```
